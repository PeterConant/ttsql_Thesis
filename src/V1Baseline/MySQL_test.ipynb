```python
# Database
from langchain_community.utilities import SQLDatabase
db = SQLDatabase.from_uri("mysql+pymysql://readonly-agent:bird@localhost:3306/bird_mini_dev")
```

```python
from langchain_core.tools import tool
import ast
from textwrap import dedent

# Define tools
@tool
def run_statement() -> str:
    """Retrieves all available tables in database by running "SHOW TABLES;" statement.
    
    Args:
        No Args
    """
    query = f"SHOW CREATE TABLE `attendance`;"
    results = db.run(query)
    
    print(results)

    return results

run_statement.invoke({})
```
